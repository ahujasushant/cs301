```python
import csv
fifa_file = open('Fifa18.csv', encoding='utf-8')
file_reader = csv.reader(fifa_file)
player_data = list(file_reader)
header = player_data[0]
player_data = player_data[1:]
for row in player_data:
    for idx in [2,6,7,8]:
        row[idx] = float(row[idx])
```

```python
def few_players():
    print(header)
    for row in player_data[:5]:
        print(row)
few_players()
```

```python
#q1
def oldest_player():
    oldest_age = 0
    for row in player_data:
        if row[2] > oldest_age:
            oldest_age = row[2]
            name = row[1]
    return name
oldest_player()
```

```python
#q2
def highest_paid_player():
    highest_salary = 0
    for row in player_data:
        if row[6] > highest_salary:
            highest_salary = row[6]
            name = row[1]
    return name
highest_paid_player()
```

```python
#q3
def highest_net_worth():
    highest_worth = 0
    for row in player_data:
        if row[7] > highest_worth:
            highest_worth = row[7]
            name = row[1]
    return name
highest_net_worth()
```

```python
#q4
def club_of_highest_net_worth():
    highest_worth = 0
    for row in player_data:
        if row[7] > highest_worth:
            highest_worth = row[7]
            club = row[4]
    return club
    
club_of_highest_net_worth()
```

```python
def get_column(col_idx):
    column_list = list()
    for row in player_data:
        column_list.append(row[col_idx])
    return column_list
get_column(0)
```

```python
#q5
def get_nationalities():
    first_five_nationalities = list()
    for nationality in get_column(3)[:5]:
        first_five_nationalities.append(nationality)
    return first_five_nationalities
get_nationalities()
```

```python
#q6 
def get_sorted_names():
    sorted_names = list()
    for name in get_column(1):
        sorted_names.append(name)
    sorted_names.sort()
    return sorted_names[:5]

get_sorted_names()
```

```python
#q7
def average_net_worth():
    total_worth = 0
    total_players = 0
    for row in player_data:
        total_worth += row[7]
        total_players += 1
    average_worth = total_worth/total_players
    return average_worth
average_net_worth()
```

```python
#q8
def average_age():
    total_age = 0
    total_players = 0
    for row in player_data:
        total_age += row[2]
        total_players += 1
    average_worth = total_age/total_players
    return average_worth

average_age()
```

```python
def player_count(country):
    players_in_country = 0
    for row in player_data:
        if row[3] == country:
            players_in_country += 1
    return players_in_country
```

```python
#q11
def highest_players_in_country():
    country_names = get_column(3)
    highest_players = 0
    for country in country_names:
        if player_count(country) > highest_players:
            highest_players = player_count(country)
            country_name = country
    return country_name

highest_players_in_country()
        
```

```python
def player_to_dict(player_id):
    player_details = list()
    for row in player_data:
        if row[0] == player_id:
            player_details = row
    return dict(zip(header, player_details))
```

```python
#q16
def players_per_nationality():
    countries_list = get_column(3)
    players_count = list()
    for country in countries_list:
        players_count.append(player_count(country))
    return dict(zip(countries_list, players_count))
players_per_nationality()    
```

```python
#q17
def players_per_league(league):
    players_count = 0
    for row in player_data:
        if row[5] == league:
            players_count += 1
    return players_count

def players_league_dict():
    leagues_list = get_column(5)
    league_players = list()
    for league in leagues_list:
        league_players.append(players_per_league(league))
    
    return dict(zip(leagues_list, league_players))
players_league_dict()
```

```python
#q18
def average_wage_per_league(league):
    total_wage_in_league = 0
    total_players_in_league = 0
    for row in player_data:
        if row[5] == league:
            total_wage_in_league += row[6]
            total_players_in_league += 1
    return total_wage_in_league/total_players_in_league

def average_wage_per_league_dict():
    leagues_list = get_column(5)
    average_wage = list()
    for league in leagues_list:
        average_wage.append(average_wage_per_league(league))

            
    return dict(zip(leagues_list, average_wage))
average_wage_per_league_dict()    
```

```python
#q19
def average_age_per_league(league):
    total_age_in_league = 0
    total_players_in_league = 0
    for row in player_data:
        if row[5] == league:
            total_age_in_league += row[2]
            total_players_in_league += 1
    return total_age_in_league/total_players_in_league

def average_age_per_league_dict():
    leagues_list = get_column(5)
    average_age = list()
    for league in leagues_list:
        average_age.append(average_age_per_league(league))

            
    return dict(zip(leagues_list, average_age))
average_age_per_league_dict()    
```

```python
#q20
def league_with_highest_average_age():
    v=list(average_age_per_league_dict().values())
    k=list(average_age_per_league_dict().keys())
    return k[v.index(max(v))]
league_with_highest_average_age()
```
